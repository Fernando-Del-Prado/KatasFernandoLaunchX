```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
```

```python
usuario_planets = input('inserta el nombre de un planeta (la primera letra en mayusculas)')
```

```python
planets_index = planets.index(usuario_planets)
```

```python
print('aqui estan los planetas mas cercanos a ' + usuario_planets)
print(planets[0:planets_index])
```

```python
print('aqui estan los planetas mas lejanos a ' +usuario_planets)
print(planets[planets_index + 1:])
```
