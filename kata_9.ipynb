```python
def promedio(valores):
    total = sum(valores)
    cantidad_de_objetos = len(valores)
    return total / cantidad_de_objetos

promedio([65, 23, 90]) 
```

```python
def reporte(tanque_1, tanque_2, tanque_3):
    return f"""combustible reporte:
    promedio total: {promedio([tanque_1, tanque_2, tanque_3])}%
    tanque 1: {tanque_1}%
    tanque 2: {tanque_2}%
    tanque 3: {tanque_3}% 
    """
    
print(reporte(30, 23, 67))
```

```python
def informe_de_mision(destino, *minutos, **reserva_de_combustible):
    return f"""reporte de mision:
    mision a {destino}
    Tiempo total de viaje: {sum(minutos)} minutos
    combustible restante: {sum(reserva_de_combustible.values())}
    """

print(informe_de_mision("marte", 10, 15, 51, main=79000, external=79000))
```

```python
def reporte_de_mision(destino, *minutos, **reservas_de_combustibles):
    principal_reporte = f"""
    Mision a {destino}
    Tiempo total de viaje: {sum(minutos)} minutos
    Combustible total restantes: {sum(reservas_de_combustibles.values())}
    """
    for nombre_del_tanque, litros in reservas_de_combustibles.items():
        principal_reporte += f"{nombre_del_tanque} tanque --> {litros} litros restantes\n"
    return principal_reporte

print(reporte_de_mision("marte", 8, 11, 55, principal=65000, externo=45000))
```
